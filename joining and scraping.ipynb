```python
import pandas as pd
import json
import requests
import os
from bs4 import BeautifulSoup
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita"
res = requests.get(url)
html=res.content
soup = BeautifulSoup(html, "html.parser")
```

```python
tables = soup.find_all("table")[1] #The second one
```

```python
df = pd.read_html(tables.prettify())[0]
df.head()
```

```python
df.keys() # we want to drop the multi index to a single index
```

```python
df.columns=df.columns.to_flat_index()
df.columns=['Country/Territory', 'UN Region', 'IMF Estimate', 'IMF Year', 'World Bank Estimate', 'World Bank Year', 'CIA Estimate', 'CIA Year' ]
list(df.columns)
#df = df.astype({('Country/Territory', 'Country/Territory'): str}, errors='raise') 
```

```python
df.columns=[(('Country/Territory', 'Country/Territory'),
            (        'UN Region',         'UN Region'),
            (    'IMF  [5]  [6]',          'Estimate'),
            (    'IMF  [5]  [6]',              'Year'),
            (  'World Bank  [7]',          'Estimate'),
            (  'World Bank  [7]',              'Year'),
            (         'CIA  [8]',          'Estimate'),
            (         'CIA  [8]',              'Year')],)]
```
